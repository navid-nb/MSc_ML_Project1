```python
import wrds

conn = wrds.Connection(wrds_username="wboughattas")
```

```python
# List all libraries
libs = conn.list_libraries()
libs
```

```python
best_libs = [i for i in libs if i.startswith(('crsp', 'comp', 'ff', 'ibes'))]
best_libs
```

```python
# List all tables in Compustat
{i: conn.list_tables(i) for i in best_libs}
```

```python
minimal_tables = {
  "crsp": [
    "dsf",                 # daily prices/returns/volume/shares
    "stocknames",          # PERMNO/TICKER name history
    "ccmxpf_linktable",    # CRSP to Compustat (PERMNO to GVKEY) with date ranges
    "dsedelist",           # delist events/returns
    "stkdistributions"     # dividends/splits (corporate actions)
  ],
  "comp": [
    "fundq",               # quarterly fundamentals (point-in-time)
    "funda",               # annual fundamentals (point-in-time)
    "company",             # GVKEY + firm header info
    "secd"                 # security master (GVKEY/IID mapping)
  ],
  "ff": [
    "factors_daily"        # Mkt-RF, SMB, HML, MOM (as available) + RF
  ],
  "ibes": [
    "actu_epsus",          # actual EPS with announcement dates (US, unadjusted)
    "statsumu_epsus",      # consensus summary history (US, unadjusted)
    "id"                   # IBES identifier master (ticker/permno/gvkey bridges)
  ]
}
```

```python
# Show variables in a table (schema)
conn.describe_table('crsp', 'dsf')
```

```python
schemas_dict = {
    f"{lib}.{tbl}": (
        (lambda df: df.to_string(index=False) if df is not None else "")(conn.describe_table(lib, tbl))
    )
    for lib, tables in minimal_tables.items()
    for tbl in tables
}
```
