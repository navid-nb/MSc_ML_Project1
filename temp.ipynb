```python
import pandas as pd
import numpy as np
dfb=pd.read_csv('temp_files\\df_prices_before_fill.csv')
dfb.set_index(['date', 'permno'], inplace=True)
dfa=pd.read_csv('temp_files\\df_prices_after_fill.csv')
dfa.set_index(['date', 'permno'], inplace=True)
def map_permno_to_ticker(df):
    # Reset index to access permno and date as columns
    df_reset = df.reset_index()

    # Check required columns existence
    if 'permno' not in df_reset.columns or 'ticker' not in df_reset.columns:
        raise ValueError("Input dataframe must contain 'permno' and 'ticker' columns in index or columns.")

    # Drop duplicates of permno and ticker pairs to get unique mappings
    unique_mappings = df_reset[['permno', 'ticker']].drop_duplicates()

    df_reset = df.reset_index() if isinstance(df.index, pd.MultiIndex) else df.copy()
    if 'permno' not in df_reset.columns:
        raise ValueError("Dataframe must have 'permno' as a column or index level.")

    grouped = df_reset.groupby('permno')
    dfs_list = [group for _, group in grouped]

    return unique_mappings,dfs_list 



print(dfb.columns)

display(dfb)
unique_mappings_b,dfs_list_b =map_permno_to_ticker(dfb)
display(unique_mappings_b)


display(dfa)
unique_mappings_a,dfs_list_a =map_permno_to_ticker(dfa)
display(unique_mappings_a)




out=dfb.copy()
out = out.groupby(level='permno').apply(lambda g: g.ffill())
out.index = out.index.droplevel(0)

unique_mappings_out,dfs_list_out =map_permno_to_ticker(out)
display(unique_mappings_out)



```
